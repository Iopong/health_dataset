# CORONA DATASET

## CONFIRMED -- DATA REQUEST

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import pytz
import tzlocal

print(pytz.utc.localize(datetime.utcnow()))
# print(tzlocal.get_localzone().localize(datetime.now()))

CONFIRMED_URL = 'https://api.covid19api.com/live/country/us/status/confirmed/date/2020-04-01T00:00:00Z'
DEATH_URL = 'https://api.covid19api.com/live/country/us/status/deaths/date/2020-04-01T00:00:00Z'
RECOVERED = 'https://api.covid19api.com/live/country/us/status/recovered/date/2020-04-01T00:00:00Z'

response = requests.get(CONFIRMED_URL)
data = response.json()
```

## VIEW DATA 

#### FIRST 5 ROWS

```python
df = pd.DataFrame(data=data)
df.head()
```

#### LAST 5 ROWS

```python
df.tail()
```

### CLEAN DATA 

#### REMOVE ANY ROWS WITH `NAN` VALUES 

```python
df = df.dropna(how='any')
df = df[df.Province != 'Recovered']
# df.info()
df
```

####  CONVERT TIME TO UTC? 

```python
time_format = "%Y-%m-%dT%H:%M:%SZ"
confirmed_date = [datetime.strptime(t, time_format) for t in df['Date']]

df['Date'] = pd.to_datetime(confirmed_date, errors='coerce').tz_localize('UTC').tz_convert('US/Pacific')
df = df.dropna(subset=['Date'])
df
```

#### GET AVERAGE CASE PER STATE

```python
confirmed_cases_df = df[['Province','Cases']].groupby('Province').mean()
confirmed_cases_df
```

### CONFIRMED - PLOT

#### BAR GRAPH

```python
confirmed_cases_df['Cases'].plot(kind='bar')
```

#### HISTOGRAM

```python
plt.figure()
confirmed_cases_df.plot.hist()
```

## DEATHS -- DATA REQUEST

```python
death_requests = requests.get(DEATH_URL)
deaths_data = death_requests.json()
```

## VIEW DATA 

```python
death_df = pd.DataFrame(data=deaths_data)
death_df.head()
```

```python
death_df.tail()
```

### CLEAN DATA

```python
death_df = death_df.dropna(how='any')
death_df = death_df[death_df.Province != 'Recovered']
death_df
```

```python
deaths_date = [datetime.strptime(t, time_format) for t in death_df['Date']]
death_df['Date'] = pd.to_datetime(deaths_date, errors='coerce').tz_localize('UTC').tz_convert('US/Pacific')
death_df = death_df.dropna(subset=['Date'])
death_df
```

### AVERAGE

```python
death_cases_df = death_df[['Province','Cases']].groupby('Province').mean()
death_cases_df
```

### GRAPHS

#### BAR GRAPH

```python
death_cases_df['Cases'].plot(kind='bar')
```

#### HISTOGRAM

```python
plt.figure()
death_cases_df.plot.hist()
```
